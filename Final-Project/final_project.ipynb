```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('hotel.csv')
df.head()
```

```python
df.info()
```

```python
#min, max, dan median dari data
df.describe()
```

```python
#min dan max dari data
count_df = df.select_dtypes(exclude=[object])
print(count_df.min())
print()
print(count_df.max())
```

```python
#drop kolom dengan 50% lebih missing value
persen = int(0.5*df.shape[0] + 1)
df = df.dropna(axis = 1, thresh = persen)
df.info()
```

```python
#mengisi missing value dengan modus
df['country'] = df['country'].fillna(df['country'].mode()[0])
df['children'] = df['children'].fillna(df['children'].mode()[0])
df['country'].mode()[0]
```

```python
#mengisi missing value dengan mean
df['agent'] = df['agent'].fillna(df['agent'].mean())
df.info()
```

```python
#Kategorikan lead time
conditions = [
    (df['lead_time'] < 30),
    (df['lead_time'] >= 30) & (df['lead_time'] <= 80),
    (df['lead_time'] > 80)
    ]

#mendefinisikan output
values = ['Short', 'Medium', 'Long']

df['Duration'] = np.select(conditions, values)
df
```

```python
#mengganti tipe data menjadi datetime
df["reservation_status_date"] = pd.to_datetime(df["reservation_status_date"],format='%Y/%m/%d', errors = 'coerce')
df.info()
```

```python
import datetime as dt

year_filter = df[df['reservation_status_date'].dt.year >= 2017]
year_filter
```

```python
#plot data count bulan
import matplotlib.pyplot as plt
%matplotlib inline

y = df["arrival_date_month"].value_counts()
x = df['arrival_date_month'].value_counts().index.tolist()

plt.barh(x, y)
plt.xlabel('Count')
plt.ylabel('Month')
plt.title('Arrival Date Month')
plt.show()
```
